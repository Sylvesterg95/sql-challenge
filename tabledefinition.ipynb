Keys
1. Primary Keys:

    *emp_no in employees: Unique employee number, making it the primary key.
    *dept_no in departments: Unique department number, making it the primary key.
*Composite primary keys for the following:
    *(emp_no, from_date) in salaries.
    *(emp_no, title, from_date) in titles.
    *(emp_no, dept_no) in dept_emp.
    *emp_no, dept_no) in dept_manager.
2. Foreign Keys:

*emp_no in dept_emp, dept_manager, titles, and salaries: References emp_no in employees.
*dept_no in dept_emp and dept_manager: References dept_no in departments.






Data Types
employees Table
1. emp_no: INT, primary key.
2. birth_date: DATE, in YYYY-MM-DD format.
3. first_name: VARCHAR, stores strings.
4. last_name: VARCHAR, stores strings.
5. gender: VARCHAR, stores one character (e.g., 'M' or 'F').
6. hire_date: DATE, in YYYY-MM-DD format.

salaries Table
1. emp_no: INT, foreign key referencing employees.
2. salary: INT, stores salary as a whole number.
3. from_date: DATE, in YYYY-MM-DD format.
4. to_date: DATE, in YYYY-MM-DD format.
titles Table

1. emp_no: INT, foreign key referencing employees.
2. title: VARCHAR, stores job titles as strings.
3. from_date: DATE, in YYYY-MM-DD format.
4. to_date: DATE, in YYYY-MM-DD format.

dept_emp Table
1. emp_no: INT, foreign key referencing employees.
2. dept_no: VARCHAR, foreign key referencing departments.
3. from_date: DATE, in YYYY-MM-DD format.
4. to_date: DATE, in YYYY-MM-DD format.

departments Table
1. dept_no: VARCHAR, primary key.
2. dept_name: VARCHAR, stores department names.

dept_manager Table
1. emp_no: INT, foreign key referencing employees.
2. dept_no: VARCHAR, foreign key referencing departments.
3. from_date: DATE, in YYYY-MM-DD format.
4. to_date: DATE, in YYYY-MM-DD format
